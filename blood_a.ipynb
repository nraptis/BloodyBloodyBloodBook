---Quick Summary of Cell Types

Basophil: Basophils are rarely seen but, when increased, suggest myeloproliferative neoplasms,
chronic inflammation, or hypersensitivity reactions.

Eosinophil: Eosinophilia most commonly points to allergic disease, parasitic infection,
drug reactions, or certain hematologic malignancies.

Erythroblast: Circulating erythroblasts indicate severe marrow stress such as hemolysis, hypoxia,
marrow infiltration, or recovery after acute blood loss.

Lymphocyte: Normal mature lymphocytes reflect routine immune surveillance, though increased
numbers suggest viral infection or lymphoproliferative disorders.

Lymphocyte (Atypical): Atypical lymphocytes are classically associated with viral infections
such as EBV, CMV, or acute immune activation.

Metamyelocyte: Presence of metamyelocytes indicates a left shift, typically due to infection,
inflammation, marrow stress, or myeloproliferative disease.

Monoblast: Monoblasts in peripheral blood are abnormal and strongly suggest acute myeloid
leukemia with monocytic differentiation.

Monocyte: Monocytosis is associated with chronic inflammation, infection, tissue repair,
autoimmune disease, or certain myeloid neoplasms.

Myeloblast: Circulating myeloblasts are pathologic and highly concerning for acute leukemia
or advanced marrow infiltration.

Myelocyte: Myelocytes in blood reflect significant marrow stimulation or dysregulation,
commonly seen in severe infection or myeloproliferative disorders.

Neutrophil (Band): Increased band neutrophils indicate an acute inflammatory response, often
referred to as a “left shift” from infection or stress.

Neutrophil (Segmented): Segmented neutrophils are normal mature granulocytes responsible for
innate immune defense against bacterial pathogens.

Promyelocyte: Promyelocytes in peripheral blood are abnormal and suggest acute leukemia
or severe marrow stress, with particular concern for APL.

Promyelocyte Bilobled: Bilobed promyelocytes are characteristic of acute promyelocytic leukemia
and carry critical diagnostic and therapeutic implications.

Smudge Cell: Smudge cells represent fragile lymphocytes and are classically associated with
chronic lymphocytic leukemia.

---NORMAL MATURE CELLS (expected baseline, context matters)

Neutrophil (Segmented):
Normal innate immune cell.

Lymphocyte:
Normal adaptive immune cell; interpretation depends on count and morphology.

---ARTIFACT-PRONE (primarily preparation-related)

Smudge Cell:
Almost always a smear-preparation artifact; becomes clinically meaningful only when numerous, where it strongly supports CLL.

---REACTIVE / STRESS-RELATED (may appear transiently)

Lymphocyte (Atypical):
Reactive immune activation, most often viral; not malignant by itself.

Neutrophil (Band):
Acute inflammatory or infectious response (“left shift”).

Metamyelocyte:
Marrow stress or inflammation; abnormal but often reactive.

Myelocyte:
Deeper left shift, usually severe infection or stress.

Eosinophil:
Commonly reactive (allergy, parasites), rarely malignant.

Basophil:
Often reactive but uncommon; persistent elevation raises suspicion for MPN.

Erythroblast:
Indicates marrow stress (hemolysis, hypoxia, bleeding) but can be reactive.

Monocyte:
Chronic inflammation, infection, or recovery states.

---PATHOLOGIC / NEOPLASTIC (should not be present in healthy adults)

Myeloblast:
Strongly indicates acute leukemia or marrow infiltration.

Monoblast:
Diagnostic for AML with monocytic differentiation when present.

Promyelocyte:
Abnormal in blood; concerning for acute leukemia.

Promyelocyte (Bilobled):
Highly specific for acute promyelocytic leukemia (APL); hematologic emergency.
